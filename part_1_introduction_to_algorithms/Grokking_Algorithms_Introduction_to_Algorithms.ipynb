```python
import matplotlib.pyplot as plt
import math
```

```python
#generating a sorted list of values

list = [i for i in range(1, 1000000)]
```

```python
# binary search function declaration

def binary_search(list, item):
  low = 0
  high = len(list) - 1  # these variables define a part of the list where function is looking for a value
  while low <= high: # function works until one value left
    mid = int((low + high) / 2) # calculating an index of the 'middle' value 
    guess = list[mid] # variable requests a value from the list using 'middle' index
    if guess == item: # if requested value equals to required one, function returns the index of this value
      return mid
    if guess > item:  # if guessed value is greater than required, function decreases highest index
      high = mid - 1
    else:             # if guessed value is lower than required, function increases lowes index
      low = mid + 1
  return None
```

```python
%%time

binary_search(list, 990000)
```

```python
def simple_search(list, item):
  index = 0
  max_index = len(list) - 1
  while index <= max_index:
    value = list[index]
    if value == item:
      return index
    else:
      index += 1
  return None
```

```python
%%time

simple_search(list, 990000)
```
